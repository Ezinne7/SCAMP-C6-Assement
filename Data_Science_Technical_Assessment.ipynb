# **Data Science Technical Assessment**


### **Import the required  libaries** 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

# Download the dataset

```python
job_posting = pd.read_csv('https://raw.githubusercontent.com/she-code-africa/SCAMP-Cohort6-Technical-Assessment/master/intermediate/fake_job_postings.csv')
```

```python
job_posting.head()
```

```python
job_posting.dtypes
```

```python
job_posting.drop(['telecommuting','has_company_logo','has_questions','salary_range','function','job_id'], axis =1, inplace = True)
```

```python
job_posting
```

**Rename the columns**

```python
job_posting = job_posting.rename(columns= {'title':'Title','location':'Location','department':'Division','company_profile':'Company_profile','description':'Job_description','requirements':'Requirement','benefits':'Benefit','employment_type':'Employment_type','required_experience':'Experience','required_education':'Educational_qualification','industry':'Industry','fraudulent':'Fraudulent'})
```

```python
job_posting
```

```python
job = job_posting.groupby(by = 'Industry', as_index = False).sum().sort_values('Fraudulent', ascending = False).reset_index(drop = True)
job
```

```python
fig= plt.figure(figsize =(20,9))
sns.barplot( x= 'Industry',y= 'Fraudulent', data = job.iloc[:10])
plt.title('The top 10 Industry with fake job posting',size = 25)
plt.xticks(rotation = 45)
plt.xlabel=('Industry')
plt.ylabel=('Fake job postings')
plt.show()
```
